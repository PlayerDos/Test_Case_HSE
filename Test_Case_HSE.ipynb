## Задача №2

### 1. Страна, в которую было совершено больше всего поездок за рубеж и количество этих поездок

```python
# подсчет поездок в страны, с сортировкой по убыванию по частоте
df_1 = df.groupby('Страна, где организована программа ')['Код работника'].count().reset_index(name='Количество поездок').sort_values(['Количество поездок'], ascending=False)
df_1.head()
```

США – страна, в которую было совершено больше всего поездок (25) за рубеж. На 2 месте Германия (11), третья Великобритания – 10 поездок.

2. Подразделение с наибольшей средней длительностью поездок сотрудников

```python
# форматирование дат, приведение к формату даты "ГГГГ-ММ-ДД"
df['Дата начала программы']  = pd.to_datetime(df['Дата начала программы'].astype(str), format='%Y%m%d %H:%M:%S.%f')
df['Дата окончания программы']  = pd.to_datetime(df['Дата окончания программы'].astype(str), format='%Y%m%d %H:%M:%S.%f')
```

```python
# добавление в исходный датафрейм столбца с количеством дней поездки
df['Длительность поездки'] = df['Дата окончания программы'] - df['Дата начала программы']
# форматирование значений в float (также прибавляем единицу для корректности значения длительности поездки)
df['Длительность поездки'] = df['Длительность поездки'].dt.days + 1
df['Длительность поездки']
```

```python
# подсчет суммарных значений длительности поездок по подразделениям
df_SumDays = df.groupby(['Код подразделения работника ВШЭ', 'Наименование подразделения'])['Длительность поездки'].sum().reset_index(name='Сумма дней')
df_SumDays.head()
```

```python
# подсчет количества поездок по подразделениям
df_ColExc = df.groupby('Код подразделения работника ВШЭ')['Код подразделения работника ВШЭ'].count().reset_index(name='Количество поездок')
df_ColExc
```

```python
# соединение таблиц в одну
df_2 = df_SumDays.merge(df_ColExc, on='Код подразделения работника ВШЭ', how='left')
df_2
```

```python
# подсчет средней продолжительности поездки по подразделениям
df_2['Средняя продолжительность'] = df_2['Сумма дней'] / df_2['Количество поездок']
# сортировка по средней продолжительности по убыванию
df_2.sort_values('Средняя продолжительность', ascending=False, inplace=True)
df_2.head()
```

Наибольшая средняя длительность поездок (35 дней) у Международной лаборатории по праву информационных технологий и интеллектуальной собственности.
Ожидаемо, подразделения с большой средней длительностью поездок имеют малое количество поездок – от 1 до 3.

3. Топ 5 подразделений московского кампуса ВШЭ по количеству участий в академической мобильности

```python
# выявление подразделений с наибольшим количеством участий в академической мобильности
df_3 = df.groupby(['Код подразделения работника ВШЭ', 'Наименование подразделения'])['Код подразделения работника ВШЭ'].count().reset_index(name='Количество поездок').sort_values(['Количество поездок'], ascending=False)
df_3(6)
```
